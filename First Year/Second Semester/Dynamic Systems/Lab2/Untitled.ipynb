```python
x = var('x')
y = function('y')(x)
a, b = var('a', 'b')
eqd = diff(y, x) == y + a * x + b
sol = desolve(eqd, [y, x], ics = [0, 1])
show(sol)
```

```python
sol1 = sol.substitute(x = 1)
sol2 = sol.substitute(x = 2)
eqd1 = sol1 == -5 + 4 * e
eqd2 = sol2 == -7 + 4 * e^2
solution = solve([eqd1, eqd2], [a, b])
show(solution)
```
