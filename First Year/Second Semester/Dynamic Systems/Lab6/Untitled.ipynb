```python
# Exercitiul 1

# a) 

t = var('t')
x = function('x')(t)

s = var('s')
f = function('f')(s)
f(s) = s^2 - 2 * s

eqp = solve(f(s) == 0, s)
eqp
```

```python
x1 = eqp[0].rhs()
x2 = eqp[1].rhs()

diff(f, s)(x1)
# -2 < 0     =>    x1 = 0 local asimptotic stabil

diff(f, s)(x2)
# 2 > 0     =>    x2 = 0 instabil
```

```python
sf = plot_slope_field(f(s), (t, -5, 5), (s, -3, 3), headaxislength = 3, headlength = 4, color = 'red')
sf
```

```python
deq = diff(x, t) == f(x)

g1 = desolve_rk4(deq, x, [0, -0.5], step = 0.1, end_points = [-5, 5], output = 'plot')
g2 = desolve_rk4(deq, x, [0, 0], step = 0.1, end_points = [-5, 5], output = 'plot')
g3 = desolve_rk4(deq, x, [0, 1], step = 0.1, end_points = [-5, 5], output = 'plot')
g4 = desolve_rk4(deq, x, [0, 2], step = 0.1, end_points = [-5, 5], output = 'plot')
g5 = desolve_rk4(deq, x, [0, 2.5], step = 0.1, end_points = [-5, 5], output = 'plot')
g = g1 + g2 + g3 + g4 + g5 + sf

g.show(ymin = -3, ymax = 3)
```

```python
# Exercitiul 2

# a) 

t = var('t')
x = function('x')(t)
y = function('y')(t)

f1(x, y) = 2 * x + y
f2(x, y) = x + 2 * y

P = desolve_system_rk4([f1(x, y), f2(x, y)], [x, y], ics = [0, 1, -1], ivar = t, end_points = [-1, 1])
XY = [ [j, k] for i, j, k in P]
XY
```

```python
A = matrix([[2, 1], [1, 2]])
A.eigenvalues()
cp(x) = A.charpoly()
solve(cp == 0, x)

# 3 > 0     =>    x1 = 3 instabil

# 1 > 0     =>    x2 = 1 instabil
```

```python
list_plot(XY, plotjoined = True, color = 'blue')
```
